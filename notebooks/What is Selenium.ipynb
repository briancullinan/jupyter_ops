```javascript
var vm = require('vm');
var path = require('path');
var fs = require('fs');
var JSONStream = require('JSONStream');
var runNotebook = ($$.runNotebook = (notebook, done) => {
    var codeCells = [];
    var counter = 0;
    var isAsync = false;
    var isDone = false;
    var runInContext = () => {
        isDone = false;
        isAsync = false;
        try {
            var temp$$ = Object.assign({}, $$, {
                done: function (msg) {
                    console.log(msg);
                    isDone = true;
                },
                async: function () {
                    isAsync = true;
                },
                ended: function () {
                    if (!isAsync) {
                        isDone = true;
                    }
                }
            })
            vm.runInNewContext(codeCells[counter])(
                exports,
                require,
                module,
                __filename,
                __dirname,
                temp$$,
                console,
                runNotebook,
                setTimeout,
                process
            );
        } catch (e) {
            console.log(e);
            isDone = true;
        }
        counter++;
    };
    var interval = setInterval(() => {
        if (isDone) {
            if(counter < codeCells.length) {
                runInContext();
            } else {
                clearInterval(interval);
                done();
            }
        }
    }, 100);
    var cells = fs.createReadStream(notebook).pipe(JSONStream.parse('cells.*'));
    cells.on('data', (cell) => {
        if(cell.cell_type === 'code') {
            var content = `
(function (
exports, 
require,
module,
__filename,
__dirname,
$$,
console,
runNotebook,
setTimeout,
process
) {
` + cell.source.join('').trim() + `
; $$.ended()})`;
            codeCells[codeCells.length] = content;
        }
    });
    cells.on('close', () => {
        isDone = true;
    });
})
$$.html(`
<script>
var base = document.getElementsByTagName('base')[0];
if (typeof base == 'undefined') {
    base = document.createElement('base');
    document.getElementsByTagName('head')[0].appendChild(base);
}
base.setAttribute('href', '\/\/` + process.cwd() + `/');
</script>
`);
```
